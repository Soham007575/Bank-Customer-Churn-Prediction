```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
import pickle
```

```python
data=pd.read_csv("Churn_Modelling.csv")
data.head()
```

# Preprocess Data

```python


data=data.drop(["RowNumber","CustomerId","Surname"] ,axis=1)
```

```python
data
```

```python
lable_encoder_gender=LabelEncoder()
data['Gender']=lable_encoder_gender.fit_transform(data['Gender'])
```

```python
data['Gender']
```

```python
data
```

```python
## Onehot encode 'Geography
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo=OneHotEncoder()
geo_encoder=onehot_encoder_geo.fit_transform(data[['Geography']]).toarray()
geo_encoder
```

```python
onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df=pd.DataFrame(geo_encoder,columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
data=pd.concat([data.drop('Geography',axis=1),geo_encoded_df],axis=1)
```

```python
data
```

# pickle file generation

```python
with open("lable_encoder_gender.pkl","wb") as file:
    pickle.dump(lable_encoder_gender,file)


with open("onehot_encoder_geo.pkl",'wb') as file:
    pickle.dump(onehot_encoder_geo,file)    
```

```python
data.head()
```

# Dependent and Independent Feature for training and testing model

```python
x=data.drop('Exited', axis=1)
y=data['Exited']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=50)

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_test
```

```python
with open('scaler.pickle','wb') as file:
    pickle.dump(scaler,file)



```

```python
data
```

# ANN implementation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
model=Sequential([

    Dense(64,activation='relu',input_shape=(x_train.shape[1],)) ,# HL1
    Dense(32,activation='relu'), # HL2
    Dense(1,activation='sigmoid') # Output
]

)
```

```python
model.summary()
```

```python
x_train.shape[1]
```

```python
opt=tf.keras.optimizers.Adam(learning_rate=0.01)
loss=tf.keras.losses.BinaryCrossentropy()
```

```python
model.compile(optimizer=opt,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Setup tensorboard

log_dir='logs/fit'+ datetime.datetime.now().strftime("%Y%m%d - %H%M%S")


```

```python
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard

log_dir='logs/fit'+ datetime.datetime.now().strftime("%Y%m%d - %H%M%S")
tensorflow_collback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
early_stopping_collback=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
history=model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks=[tensorflow_collback,early_stopping_collback]
)
```

```python
model.save('model.h5')
```
